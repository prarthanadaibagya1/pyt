Python has following data types:
1. Text type: str
2. Number type: int, float, complex
3. Sequence type: list, tuple, range
4. Mapping type: dict
5. Boolean type: bool

```python
#Basic data types of python
a=10 #int
b=99.99 #float
c=3+4j #complex
s="python" #string
l=[4,5,6] #list
t=(1,2,3) #tuple
j={1,2,3} #set
i= True #boolean
d={"Name":"Prarthana","id":28} #dict
```

```python
for i in l:
    print(i)
```

```python
for i in range (0,10,2):
    print(i)
```

List, dict, set are mutable data type while int, float, str, tuple are immutable data types.
Mutable data types can be changed after creation, changes affect the same memory location.

The type() function is used to get the type of the data

```python
print(type(a))
print(type(b))
print(type(d))
print(type(t))
```

input() takes input from the user

```python
name=input("Enter your name:")
print(f"Hello {name}")
```

The input() function takes input if=n string, so to take input in other data types we must use type casting.

```python
a= input("Enter a number")
b= input("Enter a number")
print(a+b) #if a=5, b=6 a+b gives 56
c= int(input("Enter a number"))
d= int(input("Enter a number"))
print(c+d) #if a=5, b=6 a+b gives 11
```
