List, dict, set are mutable data type while int, float, str, tuple are immutable data types.
Mutable data types can be changed after creation, changes affect the same memory location.

The type() function is used to get the type of the data

```python
print(type(a))
print(type(b))
print(type(d))
print(type(t))
```

The input() function takes input if=n string, so to take input in other data types we must use type casting.

```python
a= input("Enter a number")
b= input("Enter a number")
print(a+b) #if a=5, b=6 a+b gives 56
c= int(input("Enter a number"))
d= int(input("Enter a number"))
print(c+d) #if a=5, b=6 a+b gives 11
```
